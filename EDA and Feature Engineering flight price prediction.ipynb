## Will first feature engineer the date_of_journey

```python
df["Date"] = df["Date_of_Journey"].str.split("/").str[0]
df["Month"] = df["Date_of_Journey"].str.split("/").str[1]
df["Year"] = df["Date_of_Journey"].str.split("/").str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
## Converting the data types to int
df["Date"] = df["Date"].astype(int)
df["Month"] = df["Month"].astype(int)
df["Year"] = df["Year"].astype(int)
```

```python
df.info()
```

```python
### Dropping Date_of_journey column
df.drop("Date_of_Journey", axis = 1, inplace = True)
```

```python
df.head()
```

## Now we will feature engineer the Arrival_Time column
The seconds and month are not as important as the hour/minute so we can discard them and focus on the hour/minute
Especially because the second/month values are inconsistent 

```python
#Will remove the hour/minutes from the seconds/month
df["Arrival_Time"] = df["Arrival_Time"].apply(lambda x:x.split(" ")[0]) #The first split value will contain hour:minute
df["Arrival_hour"] = df["Arrival_Time"].str.split(":").str[0]
df["Arrival_minute"] = df["Arrival_Time"].str.split(":").str[1]
```

```python
df.head()
```

```python
df.info()
```

```python
df["Arrival_hour"] = df["Arrival_hour"].astype(int)
df["Arrival_minute"] = df["Arrival_minute"].astype(int)
df.drop("Arrival_Time", axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.head(2)
```

```python
# Feature engineering Dep_Time
df["Dep_hour"] = df["Dep_Time"].str.split(":").str[0]
df["Dep_min"] = df["Dep_Time"].str.split(":").str[1]
```

```python
df.head(2)
```

```python
df["Dep_hour"] = df["Dep_hour"].astype(int)
df["Dep_min"] = df["Dep_min"].astype(int)
df.drop("Dep_Time", axis = 1, inplace = True)
```

```python
df.head(2)
```

```python
df.info()
```

# Feature engineering Total_Stops
Contains categorical data, the more the number of stops the large the price is typically

```python
df["Total_Stops"].unique()
```

```python
df[df["Total_Stops"].isnull()]
```

```python
df["Total_Stops"].mode()
```

```python
# We can replace the missing total stop with the mode
# We will gove the stop ordinal encodings because the number of stops corresponds with higher prices
df["Total_Stops"] = df["Total_Stops"].map({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4, np.nan: 1})
```

```python
# Is now empty because we perfomed mode imputation on the NAN values
df[df["Total_Stops"].isnull()]
```

```python
df.head()
```

```python
df.info()
```

## Removing **Route**
We can remove the **Route** column because it is essentially the combination of the **Source** & **Destination** columns.

```python
df.drop("Route", axis = 1, inplace = True)
```

# Feature Engineering **Duration**

```python
# Extracting the hours
df["Duration_hours"] = df["Duration"].str.split(" ").str[0].str.split("h").str[0].fillna(0)
```

```python
# Extracting the minutes
df["Duration_min"] = df["Duration"].str.split(" ").str[1].str.split("m").str[0].fillna(0)
```
